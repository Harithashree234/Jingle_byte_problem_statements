# 5. DATA VISUALIZATION IN PYTHON

In Python, popular libraries like Matplotlib, Seaborn, and Plotly are used for creating various visualizations. These visualizations are often classified into two categories: 

 * Univariate Analysis and
 * Bivariate Analysis.

#### a) Univariate Analysis:
   Univariate analysis refers to the analysis of a single variable or feature in a dataset. The main goal of univariate analysis is to understand the distribution, central tendency, spread, and shape of the data for one feature at a time.

#### b) Bivariate Analysis:
   Bivariate analysis, on the other hand, focuses on the relationship between two variables. The aim is to explore how two variables interact with each other, whether they are correlated, and if there is any pattern or trend that links them.

## a) UNIVARIATE ANALYSIS:

### i) HISTOGRAM (Distribution of Retail Sales):

~ Histogram shows the frequency distribution of a single numerical variable by grouping data into bins. 

~ It's used to see the overall shape (e.g., normal, skewed) of the data.

```python
sns.histplot(df['RETAIL SALES'], bins=30, kde=True, color='red')
plt.title('Distribution of Retail Sales')
plt.xlabel('Retail Sales')
plt.ylabel('Frequency')
plt.show()
```

##### Chart Inferences:
* Retail sales are highly concentrated at lower values,  with significantly fewer high-value sales.
* The presence of a KDE (Kernel Density Estimate) curve indicates a right-skewed distribution.
* This means that most of the sales come from smaller transactions, which could imply a high volume of low-cost items being sold

# 

### ii) COUNT PLOT (Distribution of item types):

~ A count plot is a bar plot that shows the frequency (count) of categories in a categorical variable. 

~It is commonly used to visualize how many observations belong to each category in a dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, y='ITEM TYPE', order=df['ITEM TYPE'].value_counts().index,palette=['cyan','lime','magenta','salmon','blue','green', 'pink','indigo'])
plt.title('Distribution of Item Types')
plt.xlabel('Count')
plt.ylabel('Item Type')
plt.show()
```

##### Chart Inferences:
* The chart highlights the frequency of each item type in the dataset.
* Certain item types, such as WINE, appear more frequently than others like BEER, suggesting a skewed distribution of item types.

# 

### iii) Boxplot (Outlier Detection in Retail Sales)

A box plot (also known as a box-and-whisker plot) is a great way to visualize the distribution of a dataset and to highlight key statistics such as the median, quartiles, and outliers. 

~ The lower edge of the box represents the first quartile (Q1) (25th percentile).

~ The upper edge represents the third quartile (Q3) (75th percentile).

~ The line inside the box represents the median (50th percentile).

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='RETAIL SALES')

# Annotate the outliers (adjust xy values accordingly)
plt.annotate('Outliers', xy=(0, df['RETAIL SALES'].max() - 1), xytext=(-0.3, df['RETAIL SALES'].max() - 2),
             arrowprops=dict(facecolor='red', arrowstyle='simple'))

plt.title('Boxplot of Retail Sales (Outlier Detection)')
plt.ylabel('Retail Sales')
plt.show()
```

##### Chart Inferences:
* Outliers are clearly visible as points outside the whiskers of the boxplot.
* Most sales values fall within a relatively small range, but a few extreme values skew the dataset.

# 

### iv) PIE CHART (Proportion of Retail Sales by Top 10 Suppliers):

~ A pie chart is used to display the relative frequencies of different categories in a single variable. 

~ The size of each slice in the pie chart is proportional to the percentage of observations in each category. 

~ This means a larger slice represents a category that occurs more frequently in the dataset, while a smaller slice represents a less frequent category.

```python
# Aggregate retail sales by supplier
supplier_sales = df.groupby('SUPPLIER')['RETAIL SALES'].sum().nlargest(10)  # Top 10 suppliers by sales

# Plot the pie chart
plt.figure(figsize=(8, 8))
supplier_sales.plot.pie(
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("pastel"), 
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Proportion of Retail Sales by Top 10 Suppliers', fontsize=14)
plt.ylabel('')  # Remove y-axis label for clarity
plt.tight_layout()
plt.show()
```

##### Chart Inferences:
* The chart illustrates the proportion of retail sales by the top 10 suppliers.
* Diageo North America Inc. has the highest share at 14.8%, while Heaven Hill Distilleries Inc. and Bacardi USA Inc. both have the lowest at 6.5%.

# b) BI-VARIATE ANALYSIS:

### i) Bar Chart (Retail transfers by Item Type):

~ A bar plot (or bar chart) is a data visualization that represents categorical data with rectangular bars, where the length of each bar is proportional to the value of the category it represents. 

~It is used to compare quantities across different categories or groups.

```python
avg_sales_by_item_type = df.groupby('ITEM TYPE')['RETAIL TRANSFERS'].mean().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
avg_sales_by_item_type.plot(kind='bar', color='indigo')
plt.title('Average Retail Transfers by Item Type')
plt.ylabel('Average Retail Transfers')
plt.xlabel('Item Type')
plt.xticks(rotation=45)
plt.show()
```

##### Chart Inferences:
* Str-supplies, Non-alcohol, and Liquor have the highest average retail transfers (around 14), while DUNNAGE and KEGS have the lowest (0).
* This distribution highlights the dominance of supplies and non-alcoholic beverages in retail transfers, while other categories lag significantly behind.

### 

### ii) Box Plot ( Retail Sales by Item Type):

~ A bi-variate boxplot displays the distribution of two variables, highlighting the interquartile range, median, and potential outliers for each variable. 

~ It allows for easy comparison between the two variables to identify any correlation or differences in their spread.


```python
sns.boxplot(data=df, x='ITEM TYPE', y='RETAIL SALES')
plt.title('Retail Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Retail Sales')
plt.xticks(rotation=45)
plt.show()
```

##### Chart Inferences:
* "Beer" and "Liquor" have the highest sales in the chart, indicating a strong preference for alcoholic beverages in retail. Other categories like "Str_Supplies" and "Non-Alcohol" show significantly lower sales.
* Outliers are evident, particularly for non-alcohol items, indicating extreme values in retail sales for these categories.

# 

### iii) LINE PLOT (Monthly Retail Sales Trends):

~ A line plot is a graph that displays data points connected by straight lines, commonly used to show trends or changes over a continuous variable. 

~ It is particularly useful for visualizing the relationship between two variables, especially when one is time-dependent or ordered.


```python
df['MONTH_YEAR'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(day=1))
monthly_sales = df.groupby('MONTH_YEAR')['RETAIL SALES'].sum()
plt.figure(figsize=(8, 6))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Monthly Retail Sales Trend')
plt.ylabel('Total Retail Sales')
plt.xlabel('Month-Year')
plt.show()
```

##### Chart Inferences:
* The chart shows clear fluctuations, indicating potential seasonal trends (e.g., peaks in sales around holidays or specific months).
* Periods of low sales could represent off-peak seasons, inventory issues, or reduced consumer demand.

# 

### iv) Scatter Plot (Retail sales v/s Warehouse sales):
~ A scatter plot is a type of data visualization that displays individual data points on a two-dimensional graph, with one variable plotted on the x-axis and another on the y-axis.

~ It is used to identify relationships, trends, and correlations between two continuous variables.

```python
plt.figure(figsize=(8, 6))
sns.regplot(
    x='RETAIL SALES', y='WAREHOUSE SALES', data=df,
    scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'}
)
plt.title('Retail Sales vs Warehouse Sales')
plt.xlabel('Retail Sales')
plt.ylabel('Warehouse Sales')
plt.show()
```

##### Chart Inferences:
* A positive trendline indicates that when retail sales increase, warehouse sales also tend to rise.
* Significant spread of data points around the trendline suggests variability, possibly due to external factors like item type, supplier, or seasonal demand.
* A clustering of low sales values suggests many items may have consistently low sales in both retail and warehouse channels.
